```python
#convert video to mp3

def video2mp3(video_file, output_ext="mp3"):
    filename, ext = os.path.splitext(video_file)
    subprocess.call(["ffmpeg", "-y", "-i", video_file, f"{filename}.{output_ext}"],
                    stdout=subprocess.DEVNULL,
                    stderr=subprocess.STDOUT)
    return f"{filename}.{output_ext}"

'''
The subprocess.call method is used to run an external command and wait for its completion.
It takes a list of strings as its first argument, where each string is a
command-line argument. In this case, the list of arguments is
["ffmpeg", "-y", "-i", video_file, f"{filename}.{output_ext}"].

The -y argument to FFmpeg tells it to overwrite any existing output file with the same name.
The -i argument specifies the input file, which is passed to the function as the
video_file argument. The f"{filename}.{output_ext}" argument is a string with the
desired output file name and extension, where filename and output_ext are also
passed to the function as arguments.

The stdout argument is set to subprocess.DEVNULL, which means that the standard output
of the FFmpeg command will be discarded. The stderr argument is set to subprocess.STDOUT,
which means that the standard error of the FFmpeg command will be redirected to the same
place as the standard output.

Finally, the function returns the full output file name (f"{filename}.{output_ext}") as the result.
'''
```

```python


import gradio as gr
import os
import sys
import subprocess
#from moviepy.editor import VideoFileClip

import whisper
from whisper.utils import write_vtt

model = whisper.load_model("base")

def video2mp3(video_file, output_ext="mp3"):
    filename, ext = os.path.splitext(video_file)
    subprocess.call(["ffmpeg", "-y", "-i", video_file, f"{filename}.{output_ext}"],
                    stdout=subprocess.DEVNULL,
                    stderr=subprocess.STDOUT)
    return f"{filename}.{output_ext}"
```

```python

def translate(input_video):

    audio_file = video2mp3(input_video)

    options = dict(beam_size=5, best_of=5)
    translate_options = dict(task="translate", **options)
    result = model.transcribe(audio_file,**translate_options)

    output_dir = '/content/'
    audio_path = audio_file.split(".")[0]

    with open(os.path.join(output_dir, audio_path + ".vtt"), "w") as vtt:
      write_vtt(result["segments"], file=vtt)

    subtitle = audio_path + ".vtt"
    output_video = audio_path + "_subtitled.mp4"

    os.system(f"ffmpeg -i {input_video} -vf subtitles={subtitle} {output_video}")

    return output_video
```

```python
title = "Add Subtitles to Videos"

block = gr.Blocks()

with block:

    with gr.Group():
        with gr.Group():

            with gr.Row().style():

                inp_video = gr.Video(
                    label="Input Video",
                    type="filepath",
                    mirror_webcam=False
                )
                op_video = gr.Video()
            btn = gr.Button("Generate Subtitle Video")


    btn.click(translate, inputs=[inp_video], outputs=[op_video])
    gr.HTML('''

                    OpenAI Whisper model - Aagaman Bhattarai


    ''')
```
